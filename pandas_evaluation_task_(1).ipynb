## 1. Which month saw the highest joining of the employees

   function name: max_joinings_month()
   
   inputs: DataFrame

   outputs: str

```python
maxjoining = data.value_counts('Month Name of Joining')
maxjoining.head(1)
```

## 2. What is the mean salary is the state KS

   function name: mean_ks_salary()
   
   inputs: DataFrame

   outputs: float

```python
data[data['State']=="KS"]['Salary'].mean()
```

## 3. Which Region has highest mean salary

   function name: max_mean_salary_region()

   inputs: DataFrame

   outputs: str

```python
maxmeansalary = data.groupby('Region')['Salary'].mean()
maxmeansalary.head(1)
```

## 4. Which region has lowest mean salary

   function name: min_mean_salary_region()

   inputs: DataFrame

   outputs: str

```python
minmeansalary = data.groupby('Region')['Salary'].mean()
minmeansalary.tail(1)
```

## 5. Which City has highest mean weight

   function name: max_mean_weight_city()

   inputs: DataFrame

   outputs: str

```python
maxmeanweight = data.groupby('City')['Weight in Kgs.'].mean()
maxmeanweight
```

## 6. What is the difference between mean salary of Doctors and Professors

   function name: dr_prof_mean_salary_diff()

   inputs: DataFrame

   outputs: float

```python
dr = data[data['Name Prefix'].str.contains('Dr.')]
prof = data[data['Name Prefix'].str.contains('Prof.')]
meandr = dr['Salary'].mean()
meandr
meanprof = prof['Salary'].mean()
meanprof
print("mean of dr is: ", meandr)
print("mean of prof is: ", meanprof)
diff = meanprof - meandr
diff
```

## 7. How many female use yahoo

   function name: females_with_yahoo()

   inputs: DataFrame

   outputs: int

```python
female = data[data['Gender']=='F']
femaleswithyahoo = female[female['E Mail'].str.contains('yahoo')]
femaleswithyahoo.shape[0]
```

## 8. How many Doctors use gmail

   function name: doctors_with_gmail()

   inputs: DataFrame

   outputs: int

```python
dr = data[data['Name Prefix'].str.contains('Dr.')]
drwithgmail = dr[dr['E Mail'].str.contains('gmail')]
drwithgmail.shape[0]
```

## 9. What is the password of the youngest male

   function name: youngest_male_password()

   inputs: DataFrame

   outputs: str

```python
male = data[data['Gender']=='M']
minage = male['Age in Yrs.'].min()
minage
df_min = male[male['Age in Yrs.']== minage]
df_min

```
